```python
a=['lifetime', 'pressureInd', 'moistureInd', 'temperatureInd']
b=['broken','team', 'provider']
```

```python
for i in a:
    print(i)
    print(mydata[i].describe())
    sns.distplot(mydata[i])
    plt.show()
```

```python
for i in b:
    print(mydata[i].describe())
    sns.countplot(x=i,data=mydata)
    plt.show()
```

```python
for i in a:
    print(mydata[i].describe())
    sns.swarmplot(y=i,x='broken', data=mydata)
    plt.show()
```

```python
for i in a:
    print(mydata[i].describe())
    sns.distplot(mydata[i][mydata['broken']==0], color='red')
    sns.distplot(mydata[i][mydata['broken']==0], color='blue')
    plt.show()
```

```python
for i in a:
    print(mydata[i].describe())
    sns.boxplot(x='broken', y=i, data=mydata)
    
    plt.show()
```
